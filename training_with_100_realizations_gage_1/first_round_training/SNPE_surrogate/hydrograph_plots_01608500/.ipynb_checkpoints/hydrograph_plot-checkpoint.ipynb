## Hydrograph Plot:

```python
# ----------------------------------------------------
# Import Libraries 
# ----------------------------------------------------
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ----------------------------------------------------
# Set Base Directory and Station ID
# ----------------------------------------------------
station_id = "01608500"

# ----------------------------------------------------
# Load Observed and Baseline Data
# ----------------------------------------------------
df_obs = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_USGS.csv")
df_100_realizations = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_sim_100_sim.csv")
df_mean_std = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_mean_std.csv")
df_mean_std_100_realizations = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_100_sim+sim_mean_standard.csv")

# ----------------------------------------------------
# Convert and Align Date Columns and Drop First 5 Days
# ----------------------------------------------------
dates_obs = pd.to_datetime(df_obs["date"])[5:]
df_obs = df_obs.iloc[5:].reset_index(drop=True)

dates_100_realization = pd.to_datetime(df_100_realizations["date"])[5:]
df_100_realizations = df_100_realizations.iloc[5:].reset_index(drop=True)

dates_mean_std = pd.to_datetime(df_mean_std["date"])[5:]
df_mean_std = df_mean_std.iloc[5:].reset_index(drop=True)

dates_100_realizations_mean_std = pd.to_datetime(df_mean_std_100_realizations["date"])[5:]
df_mean_std_100_realizations = df_mean_std_100_realizations.iloc[5:].reset_index(drop=True)

# Plot observed streamflow
plt.figure(figsize=(12, 6)) 
plt.plot(dates_obs, df_obs[station_id], color='green', linewidth=2, 
         label='Observed Streamflow (USGS)', zorder=10)
plt.plot(dates_100_realization, df_100_realizations[station_id], color='blue', linewidth=2, 
         label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round:\nSNPE is Trained on Full Streamflow across 100 Realizations', 
         zorder=10)
plt.plot(dates_mean_std, df_mean_std[station_id], color='black', linewidth=2, 
         label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round: \nSNPE is trained on Summary Statistics (Mean & Std) of the Streamflow Across 100 Realizations', zorder=10)

#plt.plot(dates_mean_std, df_mean_std[station_id], color='orange', linewidth=2, label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round: \nSNPE is Trained on Summary Statistics of The Stramflow across 100 Realizations (Mean & std & Magnitude of First Peak)', zorder=10)
#plt.plot(dates_100_realizations_mean_std, df_mean_std_100_realizations[station_id], color='red', linewidth=2,
         # label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round \nTrained on Summary Statistics(Mean & Std) of the streamflow across 100 realizations and the full streamflow data',
         # zorder=10)

# Titles and labels
plt.title(f"Hydrograph Ensemble for Gage 1 ({station_id})")
plt.xlabel("Date")
plt.ylabel("Streamflow (m/hr)")
plt.legend(loc="upper left")
plt.grid(True)

# Format x-axis
locator = mdates.AutoDateLocator(minticks=6, maxticks=12)
formatter = mdates.DateFormatter('%d %b %Y')
plt.gca().xaxis.set_major_locator(locator)
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# ----------------------------------------------------
# Save the Figure
# ----------------------------------------------------
output_path = "/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/hydrograph_plots_01608500/hydrograph_training_100_realizations_mean+std_01608500.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')
print(f"[Saved] Figure saved to {output_path}")

# ----------------------------------------------------
# Show Plot
# ----------------------------------------------------
plt.show()
```

## Hydrograph Plot 

```python
# ----------------------------------------------------
# Import Libraries 
# ----------------------------------------------------
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ----------------------------------------------------
# Set Base Directory and Station ID
# ----------------------------------------------------
station_id = "01608500"

# ----------------------------------------------------
# Load Observed and Baseline Data
# ----------------------------------------------------
df_obs = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_USGS.csv")
df_100_realizations_mean_std_mag_peak = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_100_sim+mean+std+first_peak.csv")
df_mean_std_mag_first_peak = pd.read_csv("/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/streamflow_sim_excel/streamflow_daily_mean+standard+first_peak_magnitude.csv")


# ----------------------------------------------------
# Convert and Align Date Columns and Drop First 5 Days
# ----------------------------------------------------
dates_obs = pd.to_datetime(df_obs["date"])[5:]
df_obs = df_obs.iloc[5:].reset_index(drop=True)

dates_100_realizations_mean_std_mag_peak = pd.to_datetime(df_100_realizations_mean_std_mag_peak["date"])[5:]
df_100_realizations_mean_std_mag_peak = df_100_realizations_mean_std_mag_peak.iloc[5:].reset_index(drop=True)

dates_mean_std_mag_first_peak = pd.to_datetime(df_mean_std_mag_first_peak["date"])[5:]
df_mean_std_mag_first_peak = df_mean_std_mag_first_peak.iloc[5:].reset_index(drop=True)



# Plot observed streamflow
plt.figure(figsize=(12, 6)) 
plt.plot(dates_obs, df_obs[station_id], color='green', linewidth=2, label='Observed Streamflow (USGS)', zorder=10)
plt.plot(dates_100_realizations_mean_std_mag_peak, df_100_realizations_mean_std_mag_peak[station_id], color='blue', linewidth=2, 
         label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round: \nSNPE is Trained on Full Streamflow Data across 100 Realizations and their Summary Statistics (Mean & Std & Magnitude of First Peak)', zorder=10)
plt.plot(dates_mean_std, df_mean_std[station_id], color='black', linewidth=2, label='Simulated Streamflow Using Most Probable M Values From The Posterior Of The First Round: \nSNPE is Trained on Summary Statistics of The Stramflow across 100 Realizations (Mean & std & Magnitude of First Peak)', zorder=10)


# Titles and labels
plt.title(f"Hydrograph Ensemble for Gage 1 ({station_id})")
plt.xlabel("Date")
plt.ylabel("Streamflow (m/hr)")
plt.legend(loc="upper left")
plt.grid(True)

# Format x-axis
locator = mdates.AutoDateLocator(minticks=6, maxticks=12)
formatter = mdates.DateFormatter('%d %b %Y')
plt.gca().xaxis.set_major_locator(locator)
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# ----------------------------------------------------
# Save the Figure
# ----------------------------------------------------
output_path = "/home/ms6730/SBI_calibration/training _with_100_realizations_gage_1/first_round_training/SNPE_surrogate/hydrograph_plots_01608500/hydrograph_training_100_realizations_mean+std_magnitude_01608500.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')
print(f"[Saved] Figure saved to {output_path}")

# ----------------------------------------------------
# Show Plot
# ----------------------------------------------------
plt.show()
```
