```python
# ----------------------------------------------------
# Import The Libraries
# ----------------------------------------------------
import pandas as pd
import numpy as np

# ----------------------------------------------------
# Load the CSV files
# ----------------------------------------------------
file_path_sim_streamflow = "/home/ms6730/SBI_calibration/training_with_100_realizations_gage_1/first_round_training/SNL_surrogate/excel_streamflow/streamflow_daily_100_range_0.09_0.3_SNL.csv"
file_path_observed = "/home/ms6730/SBI_calibration/summary_statistics_sim_streamflow_gage_1/streamflow_excel_file/streamflow_observed_USGS.csv"

# ----------------------------------------------------
# Convert the CSV file to DataFrame
# ----------------------------------------------------
df_sim = pd.read_csv(file_path_sim_streamflow)[['date', '01608500']]
df_observed = pd.read_csv(file_path_observed)[['date', '01608500']]

# ------------------------------------------------------------------------------------------
# Exclude the first 5 rows and last row for observed (not included in training process)
# ------------------------------------------------------------------------------------------
streamflow_data_sim = df_sim['01608500'].iloc[5:]
streamflow_data_observed = df_observed['01608500'].iloc[5:-1]

# Align lengths (if needed)
min_length = min(len(streamflow_data_sim), len(streamflow_data_observed))
streamflow_data_sim = streamflow_data_sim.iloc[:min_length]
streamflow_data_observed = streamflow_data_observed.iloc[:min_length]

# ----------------------------------------------------
# Define functions for NSE, RMSE, and Bias
# ----------------------------------------------------
def nash_sutcliffe_efficiency(observed, simulated):
    return 1 - (np.sum((observed - simulated) ** 2) /
                np.sum((observed - np.mean(observed)) ** 2))

def rmse(observed, simulated):
    return np.sqrt(np.mean((observed - simulated) ** 2))

def bias(observed, simulated):
    return np.mean(simulated - observed)

# -------------------------------------
# Calculate mean and standard deviation 
# -------------------------------------
mean_value_sim_streamflow = streamflow_data_sim.mean()
std_dev_sim_streamflow = streamflow_data_sim.std()

mean_value_observed = streamflow_data_observed.mean()
std_dev_observed = streamflow_data_observed.std()

# -------------------------------------
# Calculate NSE, RMSE, and Bias
# -------------------------------------
nse_value = nash_sutcliffe_efficiency(streamflow_data_observed, streamflow_data_sim)
rmse_value = rmse(streamflow_data_observed, streamflow_data_sim)
bias_value = bias(streamflow_data_observed, streamflow_data_sim)

# ----------------------------------------------------
# Display results
# ----------------------------------------------------
print(f"Mean of the simulated streamflow (01608500): {mean_value_sim_streamflow}")
print(f"Standard deviation of simulated streamflow (01608500): {std_dev_sim_streamflow}\n")

print(f"Mean of the observed streamflow (01608500): {mean_value_observed}")
print(f"Standard deviation of observed streamflow (01608500): {std_dev_observed}\n")

print(f"Nash-Sutcliffe Efficiency (NSE): {nse_value}")
print(f"Root Mean Square Error (RMSE): {rmse_value}")
print(f"Bias: {bias_value}")
```
