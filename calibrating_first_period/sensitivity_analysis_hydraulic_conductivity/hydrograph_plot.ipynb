### Hydrograph Plot 

```python
# ----------------------------------------------------
# Import Libraries
# ----------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ----------------------------------------------------
# Set Parameters
# ----------------------------------------------------
station_id = "01608500"


#hydraulic_10_k = "/home/ms6730/SBI_calibration/calibrating_first_period/sensitivity_analysis_hydraulic_conductivity/calibrated_M_10_k.csv"
obs_file = "/home/ms6730/SBI_calibration/calibrating_first_period/sensitivity_analysis_hydraulic_conductivity/observed_streamflow_USGS.csv"
calibrated_streamflow = "/home/ms6730/SBI_calibration/calibrating_first_period/sensitivity_analysis_hydraulic_conductivity/baseline_k_calibrated_M.csv"

# ----------------------------------------------------
# Load All Data
# ----------------------------------------------------
# df_hydraulic_10_k = pd.read_csv(hydraulic_10_k)
# df_hydraulic_10_k["date"] = pd.to_datetime(df_hydraulic_10_k["date"])

df_obs = pd.read_csv(obs_file)
df_obs["date"] = pd.to_datetime(df_obs["date"])

df_calibrated = pd.read_csv(calibrated_streamflow)
df_calibrated["date"] = pd.to_datetime(df_calibrated["date"])


# ----------------------------------------------------
# Load The Dates
# ----------------------------------------------------
dates_obs = df_obs["date"]
# dates_hydraulic_10_k = df_hydraulic_10_k["date"]
dates_calibrated = df_calibrated["date"]


# ----------------------------------------------------
# Plot All Three Time Series
# ----------------------------------------------------
plt.figure(figsize=(14, 6))
plt.plot(dates_obs, df_obs[station_id], label="Observed Streamflow (USGS)", color='green', linewidth=1.8)
#plt.plot(dates_hydraulic_10_k, df_hydraulic_10_k[station_id], label="Simulated Streamflow Using Calibrated M and 10*K", color='orange', linewidth=1.8)
plt.plot(dates_calibrated,df_calibrated[station_id], label="Calibrated streamflow", color='red', linestyle='--', linewidth=1.8)

# Titles and labels
plt.title(f"Hydrograph Comparison for Station {station_id}")
plt.xlabel("Date")
plt.ylabel("Streamflow (m/hr)")
plt.legend(loc="upper left")
plt.grid(True)

# Format x-axis
locator = mdates.AutoDateLocator(minticks=6, maxticks=12)
formatter = mdates.DateFormatter('%d %b %Y')
plt.gca().xaxis.set_major_locator(locator)
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# ----------------------------------------------------
# Show Plot
# --------------
```
